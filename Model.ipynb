```python
# Convert TotalCharges to numeric, forcing any problematic values to NaN
data["TotalCharges"] = pd.to_numeric(data["TotalCharges"], errors='coerce')
data.isnull().sum()
```

```python
#totalcharges has 11 missing values, fill them with median
data["TotalCharges"] = data["TotalCharges"].fillna(data["TotalCharges"].median())
data.isnull().sum()
```

```python
#Encoding categorical variables
#binary cols has yes/no male/female
#use Label Encoding
binary_columns = ["gender", "Partner", "Dependents", "PhoneService", "PaperlessBilling", "Churn"]

for col in binary_columns:
    data[col] = data[col].replace({"Yes": 1, "No": 0, "Male": 1, "Female": 0})
```

```python
# multi-category cols has more than two possible values
#use One-Hot encoding
multi_cat_columns = ["MultipleLines", "InternetService", "OnlineSecurity", "OnlineBackup", 
                     "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies", 
                     "Contract", "PaymentMethod"]

data = pd.get_dummies(data, columns=multi_cat_columns, drop_first=True)
```
